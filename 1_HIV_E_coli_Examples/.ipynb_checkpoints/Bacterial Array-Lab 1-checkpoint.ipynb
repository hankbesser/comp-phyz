# 4.2 Bacterial Example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 1
time_constant = 1

time_bacteria = np.linspace(0,2)

W =  A * (np.exp(-time_bacteria/time_constant)-\
(1)+(time_bacteria/time_constant))

plt.plot(time_bacteria, W) 
```

```python
A = np.array([1,1.2,0.5,0.2,1.07,0.4,0.6,0.2])
time_constant = np.array([1,1.3,0.8,0.6,0.9,0.6,0.95,1.05])
```

```python
#amrount of time

W = np.zeros((8,50)) # zero array of W[i] values

time_bacteria = np.linspace(0,2)  #total time as given

W[0,0] = 0 
#assigns initial conditon (not needed in this problem 
#because starts at 0,0 point)
```

```python
#equation given
for i in np.arange(0,8):
    W[i] =  A[i] * ((np.exp(-time_bacteria/(time_constant[i])))\
                    -(1)+(time_bacteria/(time_constant[i])))
```

```python
#Plotting each line
plt.figure()

plt.plot(time_bacteria, W[0,:], "k-",linewidth=3) 
plt.plot(time_bacteria, W[1,:],"g--",linewidth=3)
plt.plot(time_bacteria, W[2,:], "b-", linewidth=2 )
plt.plot(time_bacteria, W[3,:], "m:"  ,linewidth=3 )
plt.plot(time_bacteria, W[4,:], "r--"  ,linewidth=3)
plt.plot(time_bacteria, W[5,:], "y-" ,linewidth=2 )
plt.plot(time_bacteria, W[6,:], "c-."  ,linewidth=2 )
plt.plot(time_bacteria, W[7,:], "k-.",linewidth=2 )


ax = plt.gca()

ax.legend(("W1", "W2","W3","W4","W5","W6","W7","W8"), loc=1)


    
plt.show()
```
