# Bacterial Example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 1
time_constant = 1

time_bacteria = np.linspace(0,2)

W =  A * (np.exp(-time_bacteria/time_constant)-\
(1)+(time_bacteria/time_constant))

plt.plot(time_bacteria, W) 
```

```python
A = 1
time_constant = 1

A_1 = 1.2
time_constant_1 = 1.3

A_2 = 0.5
time_constant_2 = 0.8 

A_3 = 0.2
time_constant_3 = 0.6

A_4 = 1.07
time_constant_4 = 0.9

A_5 = 0.4
time_constant_5 = 0.6

A_6 = 0.6
time_constant_6 = 0.95 

A_7 = 0.2
time_constant_7 = 1.05 

W =  A * (np.exp(-time_bacteria/time_constant)\
-(1)+(time_bacteria/time_constant))

W1 = A_1 * (np.exp(-time_bacteria/time_constant_1)\
-(1)+(time_bacteria/time_constant_1))

W2 = A_2 * (np.exp(-time_bacteria/time_constant_2)\
-(1)+(time_bacteria/time_constant_2))

W3 = A_3 * (np.exp(-time_bacteria/time_constant_3)\
-(1)+(time_bacteria/time_constant_3))

W4 = A_4 * (np.exp(-time_bacteria/time_constant_4)\
-(1)+(time_bacteria/time_constant_4))

W5 = A_5 * (np.exp(-time_bacteria/time_constant_5)\
-(1)+(time_bacteria/time_constant_5))

W6 = A_6 * (np.exp(-time_bacteria/time_constant_6)\
-(1)+(time_bacteria/time_constant_6))

W7 = A_7 * (np.exp(-time_bacteria/time_constant_7)\
-(1)+(time_bacteria/time_constant_7))


plt.figure()


plt.plot(time_bacteria, W,  linestyle=':',\
linewidth=3, color='r', label= "W")  

plt.plot(time_bacteria, W1, linestyle='--',\
linewidth=3, color='g', label= "W_1")

plt.plot(time_bacteria, W2, linestyle='-',\
linewidth=2, color='b', label= "W_2")

plt.plot(time_bacteria, W3, linestyle=':',\
linewidth=3, color='m', label= "W_3")

plt.plot(time_bacteria, W4, linestyle='--',\
linewidth=3, color='k', label= "W_4")

plt.plot(time_bacteria, W5, linestyle='-',\
linewidth=2, color='y', label= "W_5")

plt.plot(time_bacteria, W6, linestyle='-.',\
linewidth=2, color='c', label= "W_6")

plt.plot(time_bacteria, W7, linestyle='-',\
linewidth=2, color='grey', label= "W_7")

plt.legend(loc=2)
    
plt.show()
```

```python
Bacterial_data_set = np.load("novick.npz")
```

```python
Bacterial_data_set.keys()
```

```python
bac_pop_culture_A = Bacterial_data_set['g149novickA']
bac_pop_culture_A
```

```python
bac_pop_culture_B = Bacterial_data_set['g149novickB']
bac_pop_culture_B
```

```python
plt.plot(bac_pop_culture_A[:,0], bac_pop_culture_A[:,1], 'r.',\
label="Bacterial Population")       

plt.legend(loc=2)

ax = plt.gca()

ax.set_title("E. coli Bacteria Population v. Time", family='monospace',\
size=20, weight='bold')

ax.set_xlabel("Time [Hours]")
ax.set_ylabel("Population [Cells/Unit]")

ax.set_xticklabels(ax.get_xticks(), family='monospace', fontsize=10)
ax.set_yticklabels(ax.get_yticks(), family='monospace', fontsize=10)

```

```python
time_fit = np.linspace(0,7)
time_constant_fit = 3.5

V = (1) - (np.exp(-time_fit/time_constant_fit))

plt.plot(bac_pop_culture_A[:,0], bac_pop_culture_A[:,1], 'r.',\
time_fit, V, 'b-' )       


ax = plt.gca()

ax.set_title("E. coli Bacteria Population v. Time", family='monospace',\
size=16, weight='bold')

ax.set_xlabel("Time [Hours]")
ax.set_ylabel("Population [Cells/Unit]")

ax.set_xticklabels(ax.get_xticks(), family='monospace', fontsize=10)
ax.set_yticklabels(ax.get_yticks(), family='monospace', fontsize=10)

ax.legend(("Bacteria Experimental Data", "Model Fit"), loc=4)
```

```python
plt.plot(bac_pop_culture_B[0:12,0], bac_pop_culture_B[0:12,1], 'r.',\
label="Bacterial Population")
plt.legend(loc=2)

ax = plt.gca()

ax.set_title("E. coli Bacteria Population v. Time", family='monospace',\
size=20, weight='bold')

ax.set_xlabel("Time [Hours]")
ax.set_ylabel("Population [Cells/Unit]")

ax.set_xticklabels(ax.get_xticks(), family='monospace', fontsize=10)
ax.set_yticklabels(ax.get_yticks(), family='monospace', fontsize=10)
```

```python
time_fit_W = np.linspace(0,10)

A_fit = .90
time_constant_fit_W = 17.00

W_fit = A_fit * (np.exp(-time_fit_W/time_constant_fit_W)-(1)+\
(time_fit_W/time_constant_fit_W))

plt.plot(bac_pop_culture_B[0:12,0], bac_pop_culture_B[0:12,1], 'r.',\
time_fit_W, W_fit, 'b-' )       

ax = plt.gca()

ax.set_title("E. coli Bacteria Population v. Time", family='monospace',\
size=16, weight='bold')

ax.set_xlabel("Time [Hours]")
ax.set_ylabel("Population [Cells/Unit]")

ax.set_xticklabels(ax.get_xticks(), family='monospace', fontsize=10)
ax.set_yticklabels(ax.get_yticks(), family='monospace', fontsize=10)

ax.legend(("Bacterial Polulation Experimental Data", "Model Fit"), fontsize=11,\
loc=2)
```
