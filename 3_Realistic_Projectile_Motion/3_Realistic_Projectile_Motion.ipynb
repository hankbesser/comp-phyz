#### Problem under study: predicting how far a real baseball travels when hit. The exploration begins with seeing how atmospheric drag and wind effect the game of baseball.

- The force on a baseball is given by the following equation:


$$ Equation 1:  F_{drag} = -B_{2}*v^2$$

- However, real baseballs show (using wind tunnel measurements) that the drag coefficient $ B_{2} $ is actually a strong function of $ v $.

- In order to calculate the trajectory of a baseball, we need to solve the Euler equations, including the velocity dependence of the drag force. For numerical purposes, it is useful to have an approximate analytic representation for the function of the drag force (equation 1). Fortunately, the drag factor for a normal baseball is described by the following function:

$$Equation 2:  \frac{B_{2}}{m} = 0.0039 + \frac{0.0058}{1 + exp[\frac{v-v_{d}}{\Delta}]}$$
- where $v_{d} = 35 m/s$ and $\Delta = 5 m/s$

#### No wind

- The program for calculating the trajectory of a baseball, using the Euler method, and considering the velocity dependence of the drag force is described and shown below: 

```python
def baseball_no_wind(v, x0, y0, theta, time0, time_f, dt):
    """
    Calculates solution to the baseball trajectory problem using 
    the Euler method, including the 
    velocity dependence of the drag force but no wind effect. 
    """
    g = 9.8 
    
    dt = float(dt)  # time step (in seconds)
    
    N_dt = int((time_f-time0)/dt)   # number of time steps
                                    # (i.e numerical solutions)
    
    time_max = N_dt*dt # Total time (in seconds)
    
    time = np.linspace(time0, time_max, N_dt+1) # total time with 
                                            # N_dt number of time steps
    x=np.zeros(time.size) # zero array  
    y=np.zeros(time.size) # zero array 
    vel_x=np.zeros(time.size) # zero array 
    vel_y=np.zeros(time.size) # zero array 
    vel_x[0]=v*np.cos(np.radians(theta)) #initial condition
    vel_y[0]=v*np.sin(np.radians(theta))#initial condition
    x[0]=x0  #initial condition
    y[0]=y0  #initial condition
    
    for i in range(0, N_dt): 
    
        velocity = np.sqrt(vel_x[i]**2 + vel_y[i]**2)
        
        #B2/m - drag coefficent- air resistance
        drag_co = 0.0039 + (0.0058/(1+ np.exp((velocity-35)/5)))            
        
        #euler method 
        vel_x[i+1] = vel_x[i] - drag_co * velocity * vel_x[i] *  dt
        
        #euler method
        vel_y[i+1] = vel_y[i] - g*dt - drag_co * velocity * vel_y[i] *  dt 
        
        #euler method
        x[i+1] = x[i] + vel_x[i] *dt 
        
        #euler method
        y[i+1] = y[i] + vel_y[i] *dt
        
    return x,y
```

```python
# define parameters: v, x0, y0, theta, time0, time_f, dt
x1, y1 = baseball_no_wind(50, 0, 0, 35, 0, 5, 0.001)
```

#### Wind Effects

- Now Adding the effects of wind and explaining the equations:

$$ Equation 3:  F_{drag,x} = -B_{2}*|\vec{v} - {\vec{v}_{wind}}| * (v_{x}- v_{wind}) $$
$$ Equation 4:  F_{drag,y} = -B_{2}*|\vec{v} - {\vec{v}_{wind}}| * v_{y} $$

- Where ${\vec{v}_{wind}}$ is the velocity of the wind ((positive value corresponds to tailwind and negative value corresponds to headwind). 

- Transforming in to the batter's reference frame (where ${\vec{v}_{wind}} = 0$ requires to subtract the ${\vec{v}_{wind}}$ from ${\vec{v}}$, which gives Equations 3 and 4. Also, this transformation means that $B_{2}$ in Equations 3 and 4 is calculated from Equation 2 with $v$ equal to the velocity of the ball relative to the wind. 

- Modifying script for "def baseball_no_wind" to include the effect of wind: "def baseball_wind". The program uses the Euler method and is described and shown below:

```python
def baseball_wind(v, v_wind, x0, y0, theta, time0, time_f, dt):
    """
    Calculates solution to the baseball trajectory problem using 
    the Euler method, including the 
    velocity dependence of the drag force and the effect of wind. 
    """
    g = 9.8 
    
    dt = float(dt)  # time step (in seconds)
    
    N_dt = int((time_f-time0)/dt)   # number of time steps
                                    # (i.e numerical solutions)
    
    time_max = N_dt*dt # Total time (in seconds)
    
    time = np.linspace(time0, time_max, N_dt+1) # total time with 
                                            # N_dt number of time steps
    x=np.zeros(time.size) # zero array
    y=np.zeros(time.size) # zero array
    vel_x=np.zeros(time.size) # zero array
    vel_y=np.zeros(time.size) # zero array
    vel_x[0]=v*np.cos(np.radians(theta)) #initial condition 
    vel_y[0]=v*np.sin(np.radians(theta)) #initial condition
    x[0]=x0 #initial condition
    y[0]=y0 #initial condition
    
       
    
    for i in range(0, N_dt):
          
        velocity = np.sqrt(vel_x[i]**2 + vel_y[i]**2)
        
        #B2/m - drag coefficent- air resistance  
        drag_co = 0.0039 + (0.0058/(1+ np.exp((velocity-35)/5)))            
        
        # adding effects of wind
        
        #euler method
        vel_x[i+1] = vel_x[i] - drag_co * np.abs(velocity - v_wind) *\
                                                    ((vel_x[i])-v_wind) *  dt
        #euler method        
        vel_y[i+1] = vel_y[i] - g*dt - drag_co * np.abs(velocity - v_wind) *\
                                                            (vel_y[i]) *  dt
        
        
        #euler method       
        x[i+1] = x[i] + vel_x[i] *dt 
        
        #euler method
        y[i+1] = y[i] + vel_y[i] *dt
        
    return x,y
```

```python
# define parameters: v, v_wind, x0, y0, theta, time0, time_f, dt
#tailwind
x1_w1, y1_w1 =  baseball_wind(50,  4, 0, 0, 35, 0, 6, 0.001)

# define parameters: v, v_wind, x0, y0, theta, time0, time_f, dt
#headwind
x1_w2, y1_w2 =  baseball_wind(50,  -4, 0, 0, 35, 0, 5, 0.001)
```

####   Extra: Effects of Spin - Sidearm Pitch

- For a ball spinning about an axis perpendicular to the direction of travel, this speed will be
different on opposite edges of the ball. When the forces on the two sides of the ball are added,
there will be a component of the net force in the -z direction, perpendicular the velocity. This is
know as the Magnus force ($F_{M}$). To calculate the trajectory of the ball for the side-arm pitch, there
must be calculation in 3 dimension. Below is the code for the effects of spin and the code is explained.

```python
def baseball_spin(v, x0, y0, z0, theta, omega, time0, time_f, dt):
    """
    Calculates solution to the baseball trajectory problem using 
    the Euler method, including the effects of spin 
    """
    g = 9.8 
    
    dt = float(dt)  # time step (in seconds)
    
    N_dt = int((time_f-time0)/dt)   # number of time steps
                                    # (i.e numerical solutions)
    
    time_max = N_dt*dt # Total time (in seconds)
    
    time = np.linspace(time0, time_max, N_dt+1) # total time with 
                                            # N_dt number of time steps
    x=np.zeros(time.size)
    y=np.zeros(time.size)
    z=np.zeros(time.size)
    vel_x=np.zeros(time.size)
    vel_y=np.zeros(time.size)
    vel_z=np.zeros(time.size)
    vel_x[0]=v*np.cos(np.radians(theta))
    vel_y[0]=v*np.sin(np.radians(theta))
    vel_z[0]=0      
    x[0]=x0
    y[0]=y0
    z[0]=z0
       
    for i in range(0, N_dt): 
    
        velocity = np.sqrt(vel_x[i]**2 + vel_y[i]**2)
        
        #B2/m - drag coefficent- air resistance     
        drag_co = 0.0039 + (0.0058/(1+ np.exp((velocity-35)/5)))            
        
        #s0/m - spin-coefficent- effects of spin
        s0_co = 4.1e-4 # m = 149 g
         
        vel_x[i+1] = vel_x[i] - drag_co * velocity * vel_x[i] *  dt
        vel_y[i+1] = vel_y[i] - g*dt 
        
        # Important: Omega is parrallel to y-axis         
        vel_z[i+1] = vel_z[i] - s0_co * -1*(omega) * vel_x[i] * dt   
        
               
        x[i+1] = x[i] + vel_x[i] *dt 
        y[i+1] = y[i] + vel_y[i] *dt
        z[i+1] = z[i] + vel_z[i] *dt 
    return x,y,z
```

```python
# define parameters: v, x0, y0, z0, theta, omega, time0, time_f, dt
x_s, y_s, z_s = baseball_spin(30, 0, 50, 0, 0, 30, 0, 4, 0.01)
```

Now plotting all three trajectories for a batted ball (no-wind, tailwind, and headwind in the

```python
#plot trajectories until ball hits ground (i.e for y>=0)
plt.figure()
plt.plot(x1,y1,x1_w1,y1_w1,x1_w2,y1_w2)
ax = plt.gca()

ax.set_title("Trajectory of a Baseball",\
family='monospace',size=14, weight='bold')

ax.set_xlabel("x (m)")
ax.set_ylabel("y (m)")

ax.set_ylim(bottom=0)

ax.set_xticklabels(ax.get_xticks(), family='monospace', fontsize=10)
ax.set_yticklabels(ax.get_yticks(), family='monospace', fontsize=10)

ax.legend(("no wind effects", "v_wind > 0 (tailwind)",\
                    "v_wind < 0 (headwind)"),fontsize=9, loc=1)
plt.show()
```

#### Results  

The tailwind $v_{wind}$ = 4 m/s and the headwind $v_{wind}$ = -4m/s both effected the
range compared to the trajectory with of the batted ball when there is no wind. The
differences were pretty significant (+30m and -14m compared to the baseball trajectory not
effected by wind). It is essential to note that the air-resistance, which was factored into all
three trajectories, has an enormous effect. In vacuum the range is about 250m and the
trajectory is a parabola. However, including air-resistance, the trajectory is not parabolic (in
accordance to intuition). As the goal was to understand how far a real baseball should be
expected to travel when hit-- the effect of atmospheric drag was the most important effect- as
seen when simulating the trajectories using the Euler method, making it a realistic model of
the drag force. The effect of wind, also, did effect the trajectory, especially as the initial
velocity of 30 mph is much less that the initial velocity
after the ball is hit in baseball.
