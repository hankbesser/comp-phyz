#### Problem under study: 

 

$$ Equation 1: \frac{\partial^2y}{\partial t^2} =  c^2 \frac{\partial^2y}{\partial x^2} $$

This equation is known as the wave equation and is fundemental to many physics principals such as waves on a string and electromagnetic waves.

$y$ is the displacement from its equilibrium form, $x$ is the distance measured along the string, $t$ is the time, and $c$ is the parameter for the speed that the wave moves.

The main objective of this exploration is to develop numerical solutions for solving the wave equation. This will be done using algorithms to explore several aspects of wave motion: at fixed ends, free ends, and one of each.

The initial profile is assumed a "Gaussian pluck" of the string. The initial string profile is therefore:

$$ Equation 2: y_{0}(x) = exp[-k(x-x_{0})^2] $$

where k is factor that determines the width of the Gaussian Envelope and the displacement is centered at $x_{0}$

### Methods and Algorithms: Thorough Descriptions in the Code

The lab is divided into three sections: fixed ends, free ends, and one fixed/one made to oscillate. Therefore three functions are provided to calcuate and visualize string diplacement numerically at different times.

```python
def fixed_ends(dx,l):
    """ 
    Solution of wave equation for a string
    when both ends are fixed. dx is the grid spacing  
    parameter in meters, and l is length of string. Parameter for the wave speed 
    is set to 300 m/s in script. 
    
    This is a vectorised version of the code provided.
    
    There are 3 vectors containing positional information for the system
    y_new, y_current and y_previous. The vectorised version serves as a guide 
    to one--clearly understand the the algorithms-- 
    and two: provides a more effiecient way to to update
    the information.
    """
    #Define Constants
    c = 300    # wave speed in m/s
    dt = dx/c  # time in seconds
    r = c*dt/dx # dimensionless variable r to relate dt and dx to c

    n = int(l/dx) # number of grid points on the string - 1 
  
    
    #preallocate vectors for speed
    y_current=np.zeros(n+1) 
    y_previous=np.zeros(n+1)
    y_next=np.zeros(n+1) 
    
    # Initialise string position (must exclude the end points in main loop)
    for i in range(n+1):
        y_current[i]= y_previous[i] = np.exp(-1000*(i*dx - 0.3)**2)
    
    plt.figure()
    ax = plt.gca()
    ax.get_xaxis().set_ticks([])
    ax.set_xlabel("time [s]")
    ax.set_ylabel("displacement [m]")
    ax.set_title('Wave Motion Simulation on a String-With Fixed Ends:\n\
    Wave Displacement at Different Times',\
    family='monospace',size=12, weight='bold')
    ax.legend("",title='dx= %s' %(dx),loc=3) 
    for j in range(200): 
       
        #Now specify Boudary conditions-start and end boundries-
        #BECAUSE this is the fixed enpoints example    
        y_current[0]=0
        y_previous[0]=0
        y_next[0]=0
    
        y_current[-1]=0
        y_previous[-1]=0
        y_next[-1]=0
        
        for i in range(0,n-2):
            #nearest neighbor algorithm           
            y_next[i+1]=2*(1-r**2)*y_current[i+1]-\
            y_previous[i+1]+r**2*(y_current[i+2]+y_current[i])
        
        
        #creating a series of modulo operators and setting the remainder
        #eqaul to zero  to plot the generated waves
        #at specified rates at specfied generations.
        for i in range(n-2):
            y_previous[i+1]= y_current[i+1]
            y_current[i+1] = y_next[i+1]
       
        if (j%4 ==0 and j<10):
            plt.plot(y_next[:])
            
        if  (j%40 ==0 and 5<=j<=95):
            plt.plot(y_next[:])
            
        if  (j%4 ==0 and 96<=j<=105):
            plt.plot(y_next[:])
            
        if  (j%40 ==0 and 106<=j<=200):
            plt.plot(y_next[:])
        
        
            plt.plot(y_next[:])
    plt.show()
    
    #stacking the vectors to easily inpsect the object.    
    return np.vstack((y_previous,y_current,y_next))
```

```python
def free_ends(dx,l):
    """ 
    Solution of wave equation for an oscillating string
    when both ends are free. dx is the grid spacing  
    parameter in meters, and l is length of string. Parameter for the 
    wave speed, c is set to 300 m/s in script. 
    
    This is a vectorised version of the code provided.
    
    There are 3 vectors containing positional information for the system
    y_new, y_current and y_previous. The vectorised version serves as a guide 
    to one--clearly understand the the algorithms-- 
    and two: provides a more efficient way to understand the updated
    information without using temporary.
    """
    #Define Constants
    c = 300    # wave speed in m/s
    dt = dx/c  # time in seconds
    r = c*dt/dx # dimensionless variable r to relate dt and dx to c
    
    n = int(l/dx) # number of grid points on the string - 1 
      
   
    
    #preallocate vectors for speed
    y_current=np.zeros(n+1) 
    y_previous=np.zeros(n+1)
    y_next=np.zeros(n+1) 
      
    # Initialise string position (must exclude the end points in main loop)
    for i in range(n+1):
        y_current[i]= y_previous[i] = np.exp(-1000*(i*dx - 0.3)**2)
    
    plt.figure()
    ax = plt.gca()
    ax.get_xaxis().set_ticks([])
    ax.set_xlabel("time [s]")
    ax.set_ylabel("displacement [m]")
    ax.set_title("Wave Motion Simulation on a String-With Free Ends:\n\
    Wave Displacement at Different Times",\
    family='monospace',size=12, weight='bold')
    ax.legend("",title='dx= %s' %(dx),loc=3) 
    for j in range(200): 
        # ONLY DIFFERENCES from fixed end:
        #No Boundrary condtion defined, INSTEAD,
        #The positions of free ends are UPDATED to those of Nearest Neighbours
        #using the following algorith
        y_next[0]=2*(1-0.5*r**2)*y_current[0]-\
        y_previous[0]+r**2*(y_current[1])
        
        y_next[-1]=2*(1-0.5*r**2)*y_current[-1]-\
        y_previous[-1]+r**2*(y_current[-2])
        
        for i in range(n-1):
            y_next[i+1]=2*(1-r**2)*y_current[i+1]-\
            y_previous[i+1]+r**2*(y_current[i+2]+y_current[i])
       
                
        for i in range(n+1):
            #nearest neighbor algorithm (now for all n)            
            y_previous[i]= y_current[i]
            y_current[i] = y_next[i]
        
    
        #creating a series of modulo operators and setting the remainder
        #eqaul to zero to plot the generated waves
        #at specified rates at specfied generations.  
        if  (j%10 ==0 and 5<=j<=95):
            plt.plot(y_next[:])
        
        if  (j%50 ==0 and not 5<=j<=200 ):    
            plt.plot(y_next[:])
    
     #stacking the vectors to easily inpsect the object.
    return np.vstack((y_previous,y_current,y_next))
```

```python
def one_end_oscillate(amplitude,omega,dx,l):
    """ 
    Solution of wave equation for a string
    when one end if fixed and one end is free to oscillate. 
    dx is the grid spacing parameter in meters, and l is length of string. 
    Parameter for the wave speed is set to 300 m/s in script. 
    Omega, Amplitude are part of the equation designated for the movement 
    of the string element at the start (there is a fixed end)
    This equation is: y(i=0) = Amplitude*sin(omega*time) 
    
    This is a vectorised version of the code provided.
    
    There are 3 vectors containing positional information for the system
    y_new, y_current and y_previous. The vectorised version serves as a guide 
    to one--clearly understand the the algorithms-- 
    and two: provides a more effiecient way to to update
    the information.
    """
    #Define Constants
    c = 300    # wave speed in m/s
    dt = dx/c  # time in seconds
    r = c*dt/dx # dimensionless variable r to relate dt and dx to c
    
    n = int(l/dx) # number of grid points on the string - 1 
   
    
    #preallocate vectors for speed
    y_current=np.zeros(n+1) 
    y_previous=np.zeros(n+1)
    y_next=np.zeros(n+1) 
 
    # Initial Condition (excluding the fixed end boundary)
    for i in range(n+1):
        y_current[i]=y_previous[i]= amplitude* np.sin(omega*(i)*dx)
    
    #Now specify Boudary conditions- only for end boundries-
           
    y_previous[-1]=0
    y_current[-1]=0
    y_next[-1]=0
 
    
    plt.figure()
    ax = plt.gca()
    ax.get_xaxis().set_ticks([])
    ax.set_xlabel("time [s]")
    ax.set_ylabel("displacement [m]")
    ax.set_title("Wave Motion Simulation on a String-With One Free Dnd\n\
    and One End set to Oscillate: Wave Displacement at Different Times",\
    family='monospace',size=12, weight='bold')
    
   
    for j in range(500): 
       
        
        for i in range(0,n-1):
            y_next[i+1]=2*(1-r**2)*y_current[i+1]-\
            y_previous[i+1]+r**2*(y_current[i+2]+y_current[i])
       
        
        for i in range(n-1):
            #nearest neighbor algorithm (now for all n except fixed end boundary)            
            y_previous[i+1]= y_current[i+1]
            y_current[i+1] = y_next[i+1]
                
        #creating a series of modulo operators and setting the remainder
        #eqaul to zero to plot the generated waves
        #at specified rates at specfied generations.  
        if (j%50==0):
            plt.plot(y_next[:])
    
    ax.legend("",title='dx= %s and $\omega$= %s' %(dx,omega))
    return np.vstack((y_previous,y_current,y_next))
```

### Results:

#### Fixed Ends:

When the string was propagating with fixed ends, the wavepackets reflect and also invert as the strings forces allowing propagation exerts an upward/downward force on the end of the string--while the fixed end exerts an opposite force (downward/upward depending on motion of a specified wavepacket) as the wave cannot move. The amplitude of a wave does not change from after the wavepackets approached the fixed end, even though the wave inverted. When seeing more wavepackets for short increments of time (to get a better sense of the system), it is clear that at times the wavepackets come together and change the shape of the displacement of the wave that is being visualized. 

-------------------------------------------------------------------

#### Free Ends:

The wavepackets are not inverted when the wavepackets are reflected from the free ends. When viewing the behavior of the string for several plots with different time steps, the only plot that does not show drastic errors in the numerical solutions is when the time step = 0.01 -- which makes the dimensionless variable $r$ = 1. This is different from previous labs where decreasing the value of the time step usually decreased the error. In this lab decreasing or increasing the time-step from 0.01, increased the error (see figure 2). This is especially apparent when you decrease the dx value substantially, where the wavepacket grew rapidly, and the calculated string displacement diverged (clearly very inaccaurate).

Overall, as there is no force exerting an opposite force to oppose the strings force, the vertical force at the free-ends is also zero. This is seen with a spike of the wave along free ends when viewing the wave at its boundary condition. The restoring force at the ends are zero, and the wave does not become inverted.
